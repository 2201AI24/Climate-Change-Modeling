# **DATA PREPROCESSING**

```python
# --- Sea Level Data ---
sealevel['date'] = pd.to_datetime(sealevel['date'])
sealevel['year'] = sealevel['date'].dt.year
sealevel['month'] = sealevel['date'].dt.month
sea_yearly = sealevel.groupby('year', as_index=False)['mmfrom1993-2008average'].mean()

# --- Global Temperature Data ---
global_temp['dt'] = pd.to_datetime(global_temp['dt'])
global_temp['year'] = global_temp['dt'].dt.year
global_temp['month'] = global_temp['dt'].dt.month
temp_yearly = global_temp.groupby('year', as_index=False)[
    ['LandAverageTemperature', 'LandAndOceanAverageTemperature']
].mean()

# --- CO2 Data ---
CO2_df = CO2_df.rename(columns={'Year': 'year', 'Month': 'month'})
# Fill missing CO2 values using forward fill
CO2_df['Seasonally Adjusted CO2 (ppm)'] = CO2_df['Seasonally Adjusted CO2 (ppm)'].fillna(method='ffill')
CO2_yearly = CO2_df.groupby('year', as_index=False)['Seasonally Adjusted CO2 (ppm)'].mean()

# --- Merge all three ---
climate_df = pd.merge(sea_yearly, CO2_yearly, on='year', how='inner')
climate_df = pd.merge(climate_df, temp_yearly, on='year', how='inner')

# --- Final Clean Dataset ---
climate_df = climate_df.rename(columns={
    'mmfrom1993-2008average': 'SeaLevel_mm',
    'Seasonally Adjusted CO2 (ppm)': 'CO2_ppm',
    'LandAverageTemperature': 'LandTemp_C',
    'LandAndOceanAverageTemperature': 'LandOceanTemp_C'
})
```

# **Model Selection and Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Features and target
X = climate_df[['CO2_ppm']]
y = climate_df['LandOceanTemp_C']

# Time-aware split (first 80% train, last 20% test)
split_index = int(len(climate_df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Store results
results = {}

# --- Linear Regression ---
lin = LinearRegression()
lin.fit(X_train, y_train)
y_pred = lin.predict(X_test)

results['Linear Regression'] = {
    'MAE': mean_absolute_error(y_test, y_pred),
    'MSE': mean_squared_error(y_test, y_pred),
    'R2': r2_score(y_test, y_pred),
    'Adj_R2': 1 - (1-r2_score(y_test, y_pred)) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
}

# --- Polynomial Regression (deg=2,3) ---
for deg in [2,3]:
    poly = PolynomialFeatures(degree=deg)
    X_poly_train = poly.fit_transform(X_train)
    X_poly_test = poly.transform(X_test)

    model = LinearRegression()
    model.fit(X_poly_train, y_train)
    y_pred = model.predict(X_poly_test)

    results[f'Poly Regression deg={deg}'] = {
        'MAE': mean_absolute_error(y_test, y_pred),
        'MSE': mean_squared_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred),
        'Adj_R2': 1 - (1-r2_score(y_test, y_pred)) * (len(y_test)-1)/(len(y_test)-X_poly_test.shape[1]-1)
    }

# --- Ridge, Lasso, ElasticNet ---
for model_name, model in {
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=0.01),
    'ElasticNet': ElasticNet(alpha=0.01, l1_ratio=0.5)
}.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results[model_name] = {
        'MAE': mean_absolute_error(y_test, y_pred),
        'MSE': mean_squared_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred),
        'Adj_R2': 1 - (1-r2_score(y_test, y_pred)) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
    }

# --- Random Forest Regressor ---
rf = RandomForestRegressor(random_state=42, n_estimators=100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

results['RandomForest'] = {
    'MAE': mean_absolute_error(y_test, y_pred),
    'MSE': mean_squared_error(y_test, y_pred),
    'R2': r2_score(y_test, y_pred),
    'Adj_R2': 1 - (1-r2_score(y_test, y_pred)) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
}

# Convert to DataFrame for display
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd

# Define X and y
X = climate_df[['LandOceanTemp_C']]
y = climate_df['SeaLevel_mm']

# Time-aware split (80% train, 20% test)
split_index = int(len(climate_df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Store results
results_B = {}

def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    adj_r2 = 1 - (1-r2) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
    results_B[name] = [mae, mse, r2, adj_r2]

# Linear Regression
evaluate_model("Linear Regression", LinearRegression(), X_train, y_train, X_test, y_test)

# Polynomial Regression (deg=2,3)
for deg in [2, 3]:
    poly = PolynomialFeatures(degree=deg)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    lin_poly = LinearRegression()
    lin_poly.fit(X_train_poly, y_train)
    y_pred_poly = lin_poly.predict(X_test_poly)
    mae = mean_absolute_error(y_test, y_pred_poly)
    mse = mean_squared_error(y_test, y_pred_poly)
    r2 = r2_score(y_test, y_pred_poly)
    adj_r2 = 1 - (1-r2) * (len(y_test)-1)/(len(y_test)-X_test_poly.shape[1]-1)
    results_B[f"Poly Regression deg={deg}"] = [mae, mse, r2, adj_r2]

# Ridge, Lasso, ElasticNet
evaluate_model("Ridge", Ridge(), X_train, y_train, X_test, y_test)
evaluate_model("Lasso", Lasso(), X_train, y_train, X_test, y_test)
evaluate_model("ElasticNet", ElasticNet(), X_train, y_train, X_test, y_test)

# Random Forest
evaluate_model("RandomForest", RandomForestRegressor(n_estimators=100, random_state=42),
               X_train, y_train, X_test, y_test)

# Convert results to DataFrame (FIXED)
results_B_df = pd.DataFrame.from_dict(results_B, orient="index",
                                      columns=["MAE", "MSE", "R2", "Adj_R2"])
print(results_B_df)
```

```python
from sklearn.model_selection import TimeSeriesSplit

# Copy dataframe
df = climate_df.copy()

# 1. Introduce Lag Feature (5 years)
df['Temp_lag5'] = df['LandOceanTemp_C'].shift(5)

# 2. Introduce Cumulative Temperature Anomaly
df['Temp_cumsum'] = df['LandOceanTemp_C'].cumsum()

# Drop rows with NaN (from lag)
df = df.dropna()

# Define predictors for both experiments
X_lag = df[['Temp_lag5']]
X_cumsum = df[['Temp_cumsum']]
y = df['SeaLevel_mm']

# TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

def evaluate_with_cv(X, y, model):
    maes, mses, r2s = [], [], []
    for train_idx, test_idx in tscv.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        maes.append(mean_absolute_error(y_test, y_pred))
        mses.append(mean_squared_error(y_test, y_pred))
        r2s.append(r2_score(y_test, y_pred))
    return np.mean(maes), np.mean(mses), np.mean(r2s)

# Models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "ElasticNet": ElasticNet()
}

# Store results
results_lag, results_cumsum = {}, {}

for name, model in models.items():
    mae, mse, r2 = evaluate_with_cv(X_lag, y, model)
    results_lag[name] = [mae, mse, r2]

    mae, mse, r2 = evaluate_with_cv(X_cumsum, y, model)
    results_cumsum[name] = [mae, mse, r2]

# Convert to DataFrames
results_lag_df = pd.DataFrame(results_lag, index=["MAE", "MSE", "R2"]).T
results_cumsum_df = pd.DataFrame(results_cumsum, index=["MAE", "MSE", "R2"]).T

print("\n=== Lagged Temp (5 years) Results ===")
print(results_lag_df)

print("\n=== Cumulative Temp Anomaly Results ===")
print(results_cumsum_df)
```

```python
# === Model B: Temp_cumsum → Sea Level (Quadratic) ===
X_B = df[['Temp_cumsum']]   # predictor: cumulative anomaly
y_B = df['SeaLevel_mm']     # target: observed sea level

# Polynomial features
polyB = PolynomialFeatures(degree=2)
X_B_poly = polyB.fit_transform(X_B)

# Fit model
modelB = LinearRegression()
modelB.fit(X_B_poly, y_B)

# Predictions
y_pred = modelB.predict(X_B_poly)

# Evaluation
mae = mean_absolute_error(y_B, y_pred)
mse = mean_squared_error(y_B, y_pred)
r2 = r2_score(y_B, y_pred)

# Adjusted R²
n = len(y_B)          # number of samples
p = X_B_poly.shape[1] - 1  # exclude intercept
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("\n=== Quadratic Regression Results (Model B) ===")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"R2:  {r2:.4f}")
print(f"Adjusted R2: {adj_r2:.4f}")

# Scatter + fitted curve
plt.figure(figsize=(8,6))
plt.scatter(X_B, y_B, color='blue', s=20, alpha=0.6, label="Data")

# Smooth curve
X_range = np.linspace(X_B.min(), X_B.max(), 200).reshape(-1, 1)
y_range_pred = modelB.predict(polyB.transform(X_range))
plt.plot(X_range, y_range_pred, color='red', linewidth=2, label="Quadratic Fit")

plt.title("Sea Level vs Cumulative Temp Anomaly (Quadratic Fit)", fontsize=14)
plt.xlabel("Cumulative Temp Anomaly (°C)", fontsize=12)
plt.ylabel("Sea Level (mm from 1993–2008 avg)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Residuals plot
plt.figure(figsize=(8,5))
plt.scatter(X_B, y_B - y_pred, color="purple", s=20, alpha=0.6)
plt.axhline(0, color="black", linestyle="--")
plt.title("Residuals of Quadratic Regression", fontsize=14)
plt.xlabel("Cumulative Temp Anomaly (°C)", fontsize=12)
plt.ylabel("Residual (y - y_pred)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# **Prediction**

```python
def predict_sea_level_from_co2(co2_value, modelA, polyA, modelB, polyB, df):
    """
    Predict sea level rise given CO2 ppm using Model A and Model B.
    """
    import pandas as pd

    # Step 1: Predict temperature anomaly from CO2
    co2_df = pd.DataFrame([[co2_value]], columns=["CO2_ppm"])
    temp_pred = modelA.predict(polyA.transform(co2_df))[0]

    # Step 2: Compute cumulative anomaly
    last_cumsum = df['Temp_cumsum'].iloc[-1]
    temp_cumsum_pred = last_cumsum + temp_pred

    # Step 3: Predict sea level from cumulative anomaly
    temp_cumsum_df = pd.DataFrame([[temp_cumsum_pred]], columns=["Temp_cumsum"])
    sea_level_pred = modelB.predict(polyB.transform(temp_cumsum_df))[0]

    return {
        "CO2 (ppm)": co2_value,
        "Predicted Temp (°C anomaly)": float(temp_pred),
        "Predicted Cumulative Temp": float(temp_cumsum_pred),
        "Predicted Sea Level (mm)": float(sea_level_pred)
    }
```
